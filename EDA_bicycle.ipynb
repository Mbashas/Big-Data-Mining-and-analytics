```python
category_column = []
continuous_column = []

for column in bike.columns:
    if bike[column].dtype == object or len(bike[column].unique()) < 10:
        category_column.append(column)
    else:
        continuous_column.append(column)

print("categorical column:")
print(category_column)
print("\ncontinuous column:")
print(continuous_column)
```
