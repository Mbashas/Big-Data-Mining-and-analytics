```python
# Function to identify outliers using IQR method
def find_outliers(cassava, columns):
    outliers = {}
    for col in columns:
        Q1 = cassava[col].quantile(0.25)
        Q3 = cassava[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers[col] = cassava[(cassava[col] < lower_bound) | (cassava[col] > upper_bound)][col]
    return outliers

# Identify outliers in continuous columns
outliers = find_outliers(cassava, continuous_columns)

# Output columns with outliers
columns_with_outliers = [col for col, outlier_data in outliers.items() if not outlier_data.empty]
print("Columns with outliers:", columns_with_outliers)

# Display outliers for each column
for col, outlier_data in outliers.items():
    if not outlier_data.empty:
        print(f"Outliers in {col}:")
        print(outlier_data)
```

```python
def remove_outliers(columns):
    for col in columns:
        lower_quantile = cassava[col].quantile(0.25)
        upper_quantile = cassava[col].quantile(0.75)

        IQR = upper_quantile - lower_quantile

        upper_boundary = upper_quantile + 1.5 * IQR
        lower_boundary = lower_quantile - 1.5 * IQR

        cassava[col] = np.where(cassava[col] > upper_boundary, upper_boundary, cassava[col])
        cassava[col] = np.where(cassava[col] < lower_boundary, lower_boundary, cassava[col])

remove_outliers(continuous_columns)
```

```python
cassava_cat = cassava[['Sesn', 'locn', 'block', 'rep', 'tillage ', 'ferT', 'plotsize', 'HEC']]
cassava_cat
```

```python
# Rename the column to remove the space
cassava_cat = cassava_cat.rename(columns={'tillage ': 'tillage'})

# Generate a cross-tabulation between 'tillage' and 'ferT'
CrosstabResult = pd.crosstab(index=cassava_cat['plotsize'], columns=cassava_cat['ferT'])
print(CrosstabResult)
```

```python
#Generate a barplot to depict the relationship between 'tillage' and 'plotsize'
import matplotlib
%matplotlib inline
CrosstabResult.plot.bar()
```

```python
# import the required library to generate a contigency table
from scipy.stats import chi2_contingency
```

```python
# Perform Chi-square test
ChiSqResult = chi2_contingency(CrosstabResult)
```

```python

print('The P-Value of the ChiSq Test is:', ChiSqResult[1])
```

In since p > 0.05, we fail to reject the null hypothesis which can imply that there is no relationship between the two variables

# Between one categorical and continuous variables

```python
# generate a boxplot between "Plants_harvested" and "ferT"
cassava.boxplot(column='Plants_harvested', by='ferT', figsize=(18,10))
```

```python
#ANOVA tests:
# Defining a function to find the statistical relationship with all the categorical variables
def FunctionAnova(inpData, TargetVariable, CategoricalPredictorList):
    from scipy.stats import f_oneway

    # Creating an empty list of final selected predictors
    SelectedPredictors=[]
    
    print('##### ANOVA Results ##### \n')
    for predictor in CategoricalPredictorList:
        CategoryGroupLists=inpData.groupby(predictor)[TargetVariable].apply(list)
        AnovaResults = f_oneway(*CategoryGroupLists)
        
        # If the ANOVA P-Value is <0.05, that means we reject H0
        if (AnovaResults[1] < 0.05):
            print(predictor, 'is correlated with', TargetVariable, '| P-Value:', AnovaResults[1])
            SelectedPredictors.append(predictor)
        else:
            print(predictor, 'is NOT correlated with', TargetVariable, '| P-Value:', AnovaResults[1])
    
    return(SelectedPredictors)
```

```python
# Calling the function to check which categorical variables are correlated with target
CategoricalPredictorList=['ferT', 'tillage ']
FunctionAnova(inpData=cassava, 
              TargetVariable='Plants_harvested', 
              CategoricalPredictorList=CategoricalPredictorList)
```

So since It has als failed to reject the null hypothesis. The categorical predictor ferT and tillage predictor do not have a significant effect on the target variable Plants harvested.
